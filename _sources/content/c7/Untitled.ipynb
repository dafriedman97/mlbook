```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.arange(-7, 7, .01)
y = 1/(1 + np.exp(-x))

fig, ax = plt.subplots()
sns.lineplot(x, y)
ax.set_xlabel(r'$x$', size = 16)
ax.set_ylabel(r'$f(x)$', size = 16, rotation = 0, labelpad = 20)
ax.set_title('The Logistic Sigmoid Function', size = 18)
sns.despine()
```
